```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image        
```

1) Wyświetlanie obrazu wczytanego z pliku o podanej nazwie,

```python
name = input('Wpisz nazwe pliku: ') # cat.jpg
image_color = Image.open(name)
img_color = np.asarray(image_color)
plt.title('Wczytany obraz')
plt.imshow(img_color)
```

2) Sporzadzenie wykresów zmian poziomu szarosci wzdłuz wybranej linii poziomej lub pionowej o zadanej współrzednej

```python
grayscale = Image.open(name).convert("L")    #convert("L") - konwersja zdjęcia do skali szarości
gray = np.asarray(grayscale)

plt.figure(figsize=(15,10))

plt.subplot(1,2,1) 
plt.title('Obraz')                            
plt.imshow(gray, cmap='gray', vmin=0, vmax=255) #cmap - mapa kolorów, vmin i vmax - zakres, który obejmuje mapa kolorów.

plt.subplot(1,2,2)
plt.title('Histogram')
plt.hist(gray) 
plt.show() 
```

3) Wybór podobrazu (prostokatnego obszaru) o podanych współrzednych oraz jego zapis do pliku o zadanej nazwie.

```python
cropped = gray[0:200,0:300]                      #przypisanie zmiennej cropped części talibicy gray
plt.imshow(cropped, cmap='gray')
plt.savefig('cropped_image.jpg')                         
```
