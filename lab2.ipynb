#### Ćwiczenie 5.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image        
```

1) Wyświetlanie obrazu wczytanego z pliku o podanej nazwie,

```python
name = 'cat.jpg'
image_color = Image.open(name)
img_color = np.asarray(image_color)
plt.title('Wczytany obraz')
plt.imshow(img_color)
```

2) Sporzadzenie wykresów zmian poziomu szarosci wzdłuz wybranej linii poziomej lub pionowej o zadanej współrzednej

```python
grayscale = Image.open(name).convert("L")    #convert("L") - konwersja zdjęcia do skali szarości
gray = np.asarray(grayscale)

plt.figure(figsize=(15,10))

plt.subplot(1,2,1) 
plt.title('Obraz')                            
plt.imshow(gray, cmap='gray', vmin=0, vmax=255) #cmap - mapa kolorów, vmin i vmax - zakres, który obejmuje mapa kolorów.

plt.subplot(1,2,2)
plt.title('Histogram')
plt.hist(gray) 
plt.show() 
```

3) Wybór podobrazu (prostokatnego obszaru) o podanych współrzednych oraz jego zapis do pliku o zadanej nazwie.

```python
cropped = gray[0:200,0:300]                      #przypisanie zmiennej cropped części talibicy gray
plt.imshow(cropped, cmap='gray')
plt.savefig('cropped_image.jpg')                         
```

#### Ćwiczenie 7.

```python
name = 'dog.jpg'
image = Image.open(name).convert("L")
image_dark = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz')
plt.imshow(image_dark, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title("Histogram")
plt.hist(image_dark)
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title('Zbyt ciemny obraz po korekcie')
image_correction = cv.equalizeHist(image_dark)                #wyrównywanie 
plt.imshow(image_correction, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title("Histogram")
plt.hist(image_correction)
plt.show()
```
